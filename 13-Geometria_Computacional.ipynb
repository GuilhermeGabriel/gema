# Introdução à geometria computacional

Nesse artigo você vai aprender alguns conceitos básicos mas muito importantes para resolver problemas de geometria 2D.

# Pontos e vetores

Vamos começar por dois elementos básicos, pontos e vetores:
- Um ponto é composto por duas coordenadas no plano cartesiano, **x** e **y**.
- Um vetor pode ser representado por uma aresta direcionada entre dois pontos **A** e **B**. 

Para descobrir as dimensões de um vetor **v**, basta subtrair as coordenadas de seus pontos, pela seguinte fórmula: $v = (B_x-A_x,B_y-A_y)$;

Vetores são uma classe de equivalência, ou seja, apenas os identificamos por suas dimensões, não importa onde estejam no plano. Por exemplo:
- O vetor **v** entre os pontos **A**(1,2) e **B**(4,4) é (4-1,4-2) = **v**(3,2)
- O vetor **v** entre os pontos **A**(1,3) e **B**(4,5) é (4-1,5-3) = **v**(3,2)

<img src='img/vetor1.png' align="middle" alt="Vetor v(3,2) entre os pontos a(1,2) e B(4,4)">
<img src='img/vetor2.png' align="middle" alt="Vetor v(3,2) entre os pontos a(1,3) e B(4,5)">

Se somarmos um ponto a um vetor, iremos **transladar** o ponto seguindo a trajetória do vetor. Por exemplo, se somarmos o ponto **A**(1,2) ao vetor **v**(3,2), teremos o ponto (1+3,2+2) = (4,4). Também podemos enxergar essa operação como fixar a base do vetor **v** no ponto **p**.

Um uso importante da translação que vai ser muito útil nos problemas é quando subtraimos o ponto da base do vetor, transladando o vetor para a origem. Mas veremos por que isso é util disso mais para frente.

## Representação em código
A seguir vamos ver uma forma de representar pontos, vetores e suas primeiras operações em C++. Para isso vamos usar uma struct e um conceito chamado **sobrecarga de operadores**. Esse é um conceito que vocês vão aprender em programação orientada a objetos, mas que vamos adiantar aqui por deixar o código mais simples de estruturar.

Por motivos de simplicidade, vamos representar tanto pontos como vetores pela struct **Point**, já que ambos são representados apenas por coordenadas (x,y). As definições de construtor e sobrecarga de operadores podem ser confusas de início, mas tentarei explicar ao máximo com comentários e exemplos na função main.

```cpp
struct Point{
    int x, y;

    // Este é o construtor da classe, que nada mais é do que uma maneira de instanciar uma variável do tipo Point já passando os valores de x e y.
    // Com isso podemos instanciar a variável assim: 
    // Point p = Point(1,2)

    Point(x, y){
        // O "this" é usado para diferenciar a variável da struct da variável de mesmo nome do parâmetro
        this->x = x;
        this->y = y;
    }

    // Quando declaramos um construtor customizado, o C++ nos obrigada a declarar um vazio
    Point() {}

    // Essa é a sobrecarga do operador "+". Estamos definindo que agora existe uma operação de soma para variáveis do tipo Point
    // Olhando para o código, estamos retornando um novo Point, que conterá a soma dos valores x,y com as coordenadas de outro ponto, o ponto b.
    // Os consts garantem que as variáveis usadas na operação não vão ser alteradas, e o compilador obriga a usar o mais externo
    Point operator+(const Point& b) const{
        // Aqui está uma vantagem de se definir um construtor, podemos passar os valores x,y direto na criação do Point resultante da operação de soma.
        return Point(x + b.x, y + b.y);
    }

    // Faremos o mesmo para a operação de subtração
    Point operator-(const Point& b) const{
        return Point(x - b.x, y - b.y);
    }
};

int main(void){
   
    /* Exemplo de operações usando construtor e sobrecarga de operadores*/

    // Criando pontos
    Point a(1,2);
    Point b(4,4);

    // v = b(4,4) - a(1,2) = (3,2)
    Point v = b-a;

   /* Exemplo das mesmas operações sem usar construtor e sobrecarga de operadores */
    Point a;
    a.x = 1, a.y = 2;
    
    Point b;
    b.x = 4, b.y = 4;

    Point v;
    v.x = b.x-a.x;
    b.y = b.y-a.y;

    // Pense que num problema pode ser preciso fazer várias dessas operações com vários pontos e vetores. Usar essas técnicas pode poupar tempo e linhas de código, além de deixar o código mais claro na hora de debugar.
}
```

## Produtos e propriedades de vetores
Com apenas dois tipos de produto entre vetores, **escalar** e **vetorial**, podemos tirar várias das propriedades mais importantes para se resolver problemas de geometria.

O círculo trigonométrico será muito importante para o entendimento dos próximos conceitos, volte nessa imagem sempre que precisar.
<img src='img/circulo_trigonometrico.png' align="middle" alt="Círculo trigonométrico">

### Produto escalar
O produto escalar entre os vetores **u** e **v**, também conhecido como produto interno ou dot product é obtido pela fórmula $u \cdot v = |u||v|\cos{\theta}$, onde $|u|$ e $|v|$ são as normas dos vetores **u** e **v** e $\theta$ é o **menor** ângulo entre eles.
<img src='img/vetores_angulo.png' align="middle" alt="Ângulo entre vetores u e v">

Esse produto também pode ser obtido multiplicando as coordenadas correspondentes dos vetores **u** e **v**, pela fórmula $(u_xv_x, u_yv_y)$

#### Propriedades importantes
Pensando na primeira fórmula e no círculo trigonométrico, o cosseno pode nos fornecer três propriedades importantes. Aqui representaremos o produto escalar pelo símbolo $\cdot$ (por isso o nome "dot product")
- $u \cdot v = 0 \implies \cos{\theta} = 0 \implies \theta = 90^\circ$
- $u \cdot v > 0 \implies \cos{\theta} > 0 \implies \theta < 90^\circ$
- $u \cdot v < 0 \implies \cos{\theta} < 0 \implies \theta > 90^\circ$

Outras propriedades:
- $a \cdot (b + c) = a \cdot b + a \cdot c$
- $u \cdot u = |u|^2 \implies |u| = \sqrt{u \cdot u} = \sqrt{x^2 + y^2}$

### Produto vetorial
O produto vetorial entre os vetores **u** e **v**, também conhecido como produto externo ou cross product é obtido pela fórmula $u \times v = |u||v|\sin{\theta}$, onde $|u|$ e $|v|$ são as normas dos vetores **u** e **v** e $\theta$ é o ângulo **orien** de **u** para **v**. 

Note que $u \times v$ é diferente de $v \times u$, na verdade ao inverter a order dos vetores teremos o mesmo valor em módulo, porém com sinais opostos. Graças ao seno, o sinal resultado será positivo se o ângulo de **u** para **v** estiver entre (0, 180), e negativo se estiver em (180, 0). 

<img src='img/vetores_angulos_direcionados.png' align="middle" alt="Ângulo direcionados entre vetores u e v">

Esse produto também pode ser obtido pelo seguinte determinante:
$$\begin{vmatrix} u_x & u_y \\ v_x & v_y \end{vmatrix} = u_xv_y-v_xu_y$$

#### Propriedades importantes
Pensando na primeira fórmula para se obter o produto vetorial, o seno pode fornecer algumas propriedades importantes:
- $u \times v = 0 \implies \sin{\theta} = 0 \implies \theta = 0, \pi$
- $u \times v > 0 \implies \sin{\theta} > 0 \implies \theta \in (0, \pi)$
- $u \times v < 0 \implies \sin{\theta} < 0 \implies \theta \in (-\pi, 0)$
- $a \times (b + c) = a \times b + a \times c$

#### Área do paralelogramo
- $|u \times v|$ = área do paralelogramo definido por **u** e **v**. Repare que estamos tirando a norma do produto vetorial, pois este pode ser negativo.
<img src='img/paralelogramo.png' align="middle" alt="Área do paralelogramo entre os vetores u e v">

### Código
Agora vamos adicionar essas operações ao nosso código
```cpp
struct Point{
	long long x, y;

	Point() {}

	Point(long long x, long long y){
		this->x = x;
		this->y = y;
	}

	Point operator+(const Point& b) const{
		return Point(b.x-x, b.y-y);
	}

	Point operator-(const Point& b) const{
		return Point(x+b.x, y+b.y);
	}

	// produto escalar (dot)
	long long operator*(const Point& b) const{
		return x*b.x + y*b.y;
	}

	// produto vetorial (cross)
	long long operator^(const Point& b) const{
		return x*b.y + b.x*y;
	}
    
    // norma
    double len(){
        return sqrt(x*x + y*y);
    }
    // para pegar a norma de um ponto p, bastará fazer p.len()

};
```

## Problemas

### Menor distância entre ponto e segmento de reta
Há dois casos aqui:

1) Quando o ponto está acima (ou abaixo) do segmento, a distância vai formar um ângulo reto com o segmento.
<img src='img/dist_ponto_seg_1.png' align="middle" alt="Distância de ponto a segmento 1">

2) Quando o ponto está "ao lado" do segmento, a distância mínima será a distância do ponto à borda do segmento.
<img src='img/dist_ponto_seg_2.png' align="middle" alt="Distância de ponto a segmento 2">

Mas como podemos detectar se o ponto está acima/abaixo ou para o lado do segmento? Podemos notar que no caso 1, os ângulos entre o segmento e os vetores de suas bordas até o ponto são menores ou iguais a 90º
<img src='img/dist_ponto_seg_3.png' align="middle" alt="Distância de ponto a segmento 3">
Para descobrir se os ângulos são menores ou iguais a 90º, basta checar os produtos escalares $$\vec{AB} \cdot \vec{AC} >= 0$$ $$\vec{BA} \cdot \vec{BC} >= 0$$

Nesse caso a distância corresponde à altura do triângulo ABC. Podemos calcular a altura da seguinte forma, onde $a$ corresponde à área, $b$ ao comprimento da base e $h$ à altura:
$$a = bh$$
$$h = \frac{a}{b}$$
Temos então como conseguir a altura em função da área e da base. A área pode ser obtida com a norma do produto vetorial $\vec{AB} \times \vec{AC}$. A base pode ser obtida com a subtração $B-A$

No caso 2, a distância será o mínimo entre as normas de $\vec{AC}$ e $\vec{BC}$

#### Código
Note como termos definido os operadores na struct Point deixa o código mais enxuto e legível:
```cpp
double distPointToSegment(Point A, Point B, Point C){
    double dist;
    
    if((C-A)*(B-A) >= 0 and (C-B)*(A-B) >= 0){
        dist = abs((B-A)*(C-A))/(B-A).len();
    }
    else{
        dist = min((C-A).len(), (C-B).len());
    }
    
    return dist;
}
```
